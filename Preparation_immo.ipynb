```python
import pandas as pd
import numpy as np

# Charger les données depuis le fichier CSV
donnees = pd.read_csv('realtor-data.csv')

# Obtenir les noms des colonnes
noms_colonnes = donnees.columns.tolist()

# Calculer la proportion de valeurs vides pour chaque colonne
proportion_valeurs_vides = donnees.isnull().mean()

# Afficher les résultats
print("Noms des colonnes : ", noms_colonnes)
print("\nProportion de valeurs vides par colonne :\n", proportion_valeurs_vides)
```

```python
# Compter le nombre total de lignes avant la suppression
nombre_lignes_avant = donnees.shape[0]
print(f"Nombre de lignes avant la suppression : {nombre_lignes_avant}")
donnees_sans_duplicat=donnees.drop_duplicates()
donnees_sans_valeurs_manquantes = donnees_sans_duplicat.dropna()
nombre_lignes_apres = donnees_sans_valeurs_manquantes.shape[0]
print(f"Nombre de lignes après la suppression : {nombre_lignes_apres}")

proportions = donnees_sans_valeurs_manquantes['status'].value_counts(normalize=True)
print(proportions)

immo = donnees_sans_valeurs_manquantes
immo.to_csv('immo.csv', index=False)
print(immo.shape)
print(immo.describe())
print(immo.dtypes)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

immo = pd.read_csv('immo.csv')

# Visualisation de la colonne 'price'

# Histogramme
plt.figure(figsize=(10, 6))
plt.hist(immo['price'], bins=50, color='blue', edgecolor='black')
plt.title('Histogramme des Prix')
plt.xlabel('Prix')
plt.ylabel('Fréquence')
plt.show()

# Boîte à moustaches
plt.figure(figsize=(6, 3))
plt.boxplot(immo['price'], vert=False)
plt.title('Boîte à Moustaches des Prix')
plt.xlabel('Prix')
plt.show()
```
